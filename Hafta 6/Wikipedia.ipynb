```python
import wikipediaapi
```

```python
wiki_pedia = wikipediaapi.Wikipedia(user_agent='MyApp/1.0 (Contact: myemail@example.com)', language='tr')
```

```python
page = wiki_pedia.page('Türkiye').text
```

```python
page
```

```python
# Bu metni kullanarak word2vec modeli oluşturmanızı istiyorum.
```

```python
import re
sentences = [re.findall(r'\b\w+\b', i.lower()) for i in page.split(".")]
```

```python
pip install gensim
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences, vector_size=50, window=5, min_count=2, workers=4, epochs=50)
```

```python
model.wv.most_similar('türkiye', topn=3)
```

```python
model.wv["türkiye"]
```
